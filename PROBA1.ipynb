```python
"""
MATHIAM FAYE 
THILOR SENE
Fatou DIARRA
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def estime(N,n):
    p=[]                                                #initialisation d'un tableau P
    for k in range(N):                                   #TIRAGE des réalisations {y1,...,yn} des n variables {Y1,...,Yn}
        y=np.random.uniform(low=0, high=10, size=[n,1])    #calcule d'une realisation de la moyenne empirique
        Y=y.sum()/n                                         #a chaque tirage,on conserve la moyenne obtenue dans le tableau P
        p.append(Y)
    return p

tango=estime(5000,10)         #on appelle la fonction estime pour N=5000
plt.figure()                  #on construit l'histogramme 
plt.title('histogramme de des 5 000 réalisations de la moyenne empirique')
plt.hist(tango)
  
iters=[10,100,1000,10000]        #on met les tailles d'echantillons dans un tableau qu'on va iterer avec l'indiçage pour recuperer n
ity=np.array(iters)              #mettons le dans un tableau numpy np.array()
N=10000                          #fixons le nombre fois que nous voulons calculer la moyenne 
for i in range(4):               #recuperation des n avec une boucle for 
    n=ity[i]                     #a chaque tour de boucle on recupere une valeure de n
    z=estime(N,n)                #la fonction estime(N,n) s'execute N fois 
    plt.figure()
    plt.title('histogramme des 10 000 réalisations avec un echatillon de taille n')
    plt.hist(z)                 #et puis l'histogramme des moyennes est traçé pour un echantillon de taille n
""" l'histogramme des 10000 realisation de la moyenne avec une echantillon de taille n=10 se rapproche plus de celui 5000 realisations de la moyenne empiruqe qui est similaire à la courbe de densite de la loi normale d'esperence 5"""
""" les allures des histogrammes se rapprochent fortement de la courbe de densité de la loi normale et plus la taille de l'echantillon est grande plus les valeurs se concentrent au tour de la moyenne qui est de 5 donc la moyenne empirique de ces variables converge en probabilité vers l’espérance m=5  """
```
