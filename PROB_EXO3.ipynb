```python
import string 
import urllib.request
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections

L =urllib.request.urlopen('https://r-stat-sc-donnees.github.io/LesMiserables1.txt')
livre=str(L.read()) 


#Suppression de la ponctuation
LesMiserables = livre.translate(str.maketrans("","", string.punctuation))
#print(LesMiserables)

#Transformation en majuscule de tous les caracteres minuscules
lesmiserables=LesMiserables.lower()
#print(lesmiserables)

# division du texte en jetons individuels (mots).
table=lesmiserables.split()
print(table)

#calcul du nombre de mots (un, et, le, il, est) respectifs au fur et à mesure que le livre progresse.
Nun=0
for i in range(len(table)):
    if table[i] == "un":
        Nun = Nun + 1
un=Nun

Net=0     # initialisation du nombre de et
for i in range(len(table)):     #on parcourt tous les mots du livre avec une boucle
    if table[i] == "et":
        Net = Net + 1
et=Net

Nle=0
for i in range(len(table)):
    if table[i] == "le":
        Nle = Nle + 1
le=Nle

Nil=0
for i in range(len(table)):
    if table[i] == "il":
        Nil = Nil + 1
il=Nil

Nest=0
for i in range(len(table)):
    if table[i] == "et":
        Nest = Nest + 1
est=Nest
print("Le nombre de un est:",un)
print("Le nombre de et est:",et)
print("Le nombre de le est:",le)
print("Le nombre de il est:",il)
print("Le nombre de est est:",est)

#compter les nombre d'occurence de chaque mot du livre
dt=dict()
for w in table:
    dt[w]=dt.get(w,0)+1
totale=collections.Counter(dt)   #definition d'un comptereur qui renvoi le nombre total de chaque mot du livre 
print(totale)
print(min(totale.values()))
print(max(totale.values()))
print(len(totale))
#crayons un dictionnaire pour des mots et de leur nombre d'occurence
#cherchons le mot le plus frequent du livre et  de son nombre d'occurence .
largest=-1
lemot=None
mot=[]
nombre=[]
for k,v in dt.items():
#    print(k,v)
    if v>largest:
        largest=v
        lemot=k
    mot.append(k)
    nombre.append(v)
#print(lemot, largest)    #renvoi le mot le plus frequent du livre et le nombre total de fois qu'il a ete ecrit
#print(mot)   #renvoi une liste de touts les mots du livre
#print(nombre) #renvoi une liste du nombre de repetition des mots du livre
N=len(mot)   #le nombre total des mots du livre
print(N)
moyenne=np.mean(nombre)
ecartype=np.std(nombre)
variance=np.var(nombre)
print(moyenne)
print(ecartype)
print(variance)

def normale(x, mu, sigma):
    sqrt_two_pi= np.sqrt(2*np.pi)
    return (np.exp(-(x-mu)**2/2/sigma**2)/(sqrt_two_pi*sigma))
plt.plot([k for k in totale.keys()], [v/N for v in totale.values()], 0.00000001, color="blue")
plt.title("Proportion de mot du livre")
mu=moyenne
sigma=ecartype
xs=range(np.min(nombre), np.max(nombre))
ys=[normale(i,mu,sigma) for i in xs] 
plt.plot(xs,ys)
plt.show()

"""
La distribution des mots du livre ne suit aucune loi, elle est totalemement aleatoire
Pour appliquer le theoreme centrale limite au mots du livre, procedons par echantillonnage.
exemple:par page
et pour chaque page, on calcule le nombre moyen d'occurence de mots
si on on effectue ce processus pour toutes les page du livre, et remarque
que les moyennes des echantillion(moyenne d'occurence des mots par page de livre
ont tendance a ce rapprocher de la moyenne d'occurence des mots du livre.
car les valeurs extremes s'equilibrent. la moyenne d'echantillion est 
une approximation de la moyenne des mots du livre.
on choisit les pages où le nombre de mot est superieur à 30
 
dans ce cas de figure, le theoreme centrale limite pourra etre appliquer 
et on verra que les moyennes suivront une loi normale
"""   
```
